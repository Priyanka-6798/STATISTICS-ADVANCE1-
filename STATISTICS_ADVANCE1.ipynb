## STATISTICS ADVANCE-1

1. What is a random variable in probability theory?

ANS: A random variable is a numerical quantity whose value depends on the outcomes of a random phenomenon. It maps outcomes of a random process to real numbers.

2. What are the types of random variables?

ANS: *Discrete Random Variables: Take countable values (e.g., number of heads in coin tosses).

  *Continuous Random Variables: Take uncountable, infinite values within an interval (e.g., height, weight).

3. What is the difference between discrete and continuous distributions?

ANS: *Discrete Distributions: Describe probabilities for countable outcomes (e.g., Binomial, Poisson).

  *Continuous Distributions: Describe probabilities over a range of values using probability density functions (e.g., Normal, Uniform).

4. What are probability distribution functions (PDF)?

ANS: *For Discrete Variables: PDF (or PMF, Probability Mass Function) gives the probability of each possible value.

   *For Continuous Variables: PDF describes the relative likelihood of a range of values (area under the curve gives probability).

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

ANS: *PDF: Provides probabilities for specific values (discrete) or densities (continuous).

   *CDF: Gives the probability that a random variable is less than or equal to a value, i.e.,
               F(x)=P(X≤x).

6. What is a discrete uniform distribution?

ANS: A distribution where each outcome in a finite set has equal probability (e.g., rolling a fair die:
                 P(X=k)=1/6 for k=1,2,.....6.

7. What are the key properties of a Bernoulli distribution?

ANS: *Models a single trial with two outcomes: success (1) with probability
p failure (0) with probability 1-p.
     *Mean=p, Variance=p(1-p).

8. What is the binomial distribution, and how is it used in probability?

ANS: *Describes the number of successes k in n independent Bernoulli trials.
     *Used in scenarios like quality control (defective items) or survey responses.

9. What is the Poisson distribution and where is it applied?

ANS: *Models rare events over a fixed interval/time (e.g., calls at a call center).

  *Assumptions: Events are independent and occur at a constant rate.

10. What is a continuous uniform distribution?

ANS: *A distribution where all values in an interval[a,b] are equally likely.
     PDF= f(x)=  1/(b-a) for a<x<b.

11. What are the characteristics of a normal distribution?

ANS: *Symmetric, bell-shaped curve.

   *68-95-99.7 Rule: ~68% of data within μ±σ, ~95% within μ±2σ.

   *Defined by mean μ (center) and standard deviation σ (spread).

12. What is the standard normal distribution, and why is it important?

ANS: *A normal distribution with μ=0 and σ=1.

  *Converts any normal variable X to Z-scores: X−μ/σ.

  *Basis for statistical tables and hypothesis testing.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

ANS: *CLT states that the sampling distribution of the mean of any independent, identically distributed (i.i.d.) variables approaches normality as n→∞, regardless of the population distribution.

 *Enables use of normal approximations for inference (e.g., confidence intervals).

14. How does the Central Limit Theorem relate to the normal distribution?

ANS: CLT justifies why many real-world processes (e.g., sample means) can be modeled using the normal distribution, even if underlying data is non-normal.

15. What is the application of Z statistics in hypothesis testing?

ANS: Z-tests compare sample statistics to population parameters using the standard normal distribution (e.g., testing if a sample mean differs from a known population mean).

16. How do you calculate a Z-score, and what does it represent?

ANS: *Formula: Z= X−μ/σ.

   *It measures how many standard deviations X is from the mean μ.

17. What are point estimates and interval estimates in statistics?

ANS: *Point Estimate: Single value approximating a parameter (e.g., sample mean x for population mean μ).

  *Interval Estimate: Range (e.g., confidence interval) providing a precision measure.

18. What is the significance of confidence intervals in statistical analysis?

ANS: They provide a range of plausible values for a parameter (e.g., population mean) with a specified confidence level (e.g., 95%).

19. What is the relationship between a Z-score and a confidence interval?
For a normal distribution, a 95% CI for the mean is:
x±Z α/2, where Z α/2=1.96.

20. How are Z-scores used to compare different distributions?

ANS: By converting values to Z-scores, we standardize different scales, enabling comparison across datasets (e.g., test scores from different exams).

21. What are the assumptions for applying the Central Limit Theorem?

ANS: *Random sampling.

   *Independent observations.

   *Sample size n≥30 (for non-normal populations).

22. What is the concept of expected value in a probability distribution?

ANS: The long-run average value of repetitions of an experiment (e.g., E(X)=∑x⋅P(X=x) for discrete variables).

23. How does a probability distribution relate to the expected outcome of a random variable?

ANS: The expected value E(X) is the theoretical mean of the distribution, summarizing the center of the variable’s possible outcomes.
